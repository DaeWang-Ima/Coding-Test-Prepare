# 배열

- 메모리 상에 원소를 연속하게 배치한 자료구조

# 배열의 성질

- o(1)에 k번째 원소를 확인/변경 가능

- 추가적으로 소모되는 메모리의 양(overhead)이 거의 없다.

- cache hit rate가 높다.

- 메모리 상에 연속한 구간을 잡아야 해서 할당에 제약이 걸린다.

# 배열에서 제공하는 연산

- 임의의 위치에 있는 원소를 확인/변경 : O(1)

- 원소를 끝에 추가 : O(1) → 길이를 1 증가시키면 가능

- 마지막 원소를 제거 : O(1) → 길이를 줄이면 가능

- 임의의 위치에 있는 원소를 추가 : O(n) → 중간에 원소를 추가하면, 그 뒤에 원소들을 전부 한 칸씩 밀어야 가능

    - 예를 들어 책장에 책이 연속해서 꽂혀있을 때, 중간에 책을 넣기 위해서는 나머지 책들을 밀어야하는 상황이 있다고 하자. 추가하는 위치가 끝에 가까울수록 시간을 줄어들고 앞에 가까울수록 시간은 늘어나지만, 평균적으로는 n/2 개를 밀어야하므로, O(n) 이라고 생각하면 된다.<br><br>
    
- 임의의 위치에 있는 원소를 제거 : O(n) → 5번 원소를 지우면, 그 이후에 있는 모든 원소들을 한 칸씩 땡겨와야한다.

    - 5번 원소만 비워두고 다른 원소들은 그대로 두면 안될까?<br><br>
    
    - 메모리 상에 원소가 연속해서 존재하지 않기 때문에 배열의 정의에도 위배되고, k번째 원소를 O(1)에 찾을수가 없게 된다.

# 배열을 사용하는 기본 알고리즘

## 1. 배열 원소의 최댓값 구하기

- 만약 a = [10, 54, 14, 88, 43] 이라는 배열이 있다고 가정하자. 여기서 최댓값을 구하는 과정은 다음과 같다.

- 1. a[0]의 값을 maximum에 대입한다.<br><br>

- 2. a[1] ~ a[4] 까지 차례대로 비교한다.<br><br>

- 3. a[i]가 maximum보다 크면 maximum 값에 대입한다.<br><br>

- 4. 배열의 모든 원소에 대해 스캔을 마치면, 배열 a의 최댓값이 maximum에 남아있게 된다.

```python
from typing import Any, Sequence

# Any : 제약이 없는 임의의 자료형
# Sequence : 시퀀스형 (리스트형, 튜플형, 문자형 등...)
```

```python
def find_max(a: Sequence) -> Any:
    
    # 초기값 설정
    maximum = a[0]
    
    for i in range(1, len(a)):
        
        if a[i] > maximum:
            maximum = a[i]
    
    return maximum
```

```python
if __name__ == "__main__":
    print("배열의 최댓값 구하기")
    
    num = int(input("원소의 개수 : "))
    
    x = [None] * num
    
    for i in range(num):
        x[i] = int(input(f"x[{i}] 값을 입력하세요. "))
        
    print(f"배열의 최댓값은 {find_max(x)}입니다.")
```
